## Write a Python Program using Perceptron Neural Network to recognise even and odd numbers. Given numbers are in ASCII form 0 to 9

```python
import numpy as np
```

```python
class Perceptron(object):
    def __init__(self, no_of_inputs, threshold=100, learning_rate=0.01):
        self.threshold = threshold
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_inputs + 1)
        
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation > 0:
            activation = 1
        else:
            activation = 0            
        return activation

    def train(self, training_inputs, labels):
        for _ in range(self.threshold):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] += self.learning_rate * (label - prediction) * np.array(inputs)
                self.weights[0] += self.learning_rate * (label - prediction)
```

```python
training_inputs = []
labels = []
```

```python
for i in range(10):
    training_inputs.append([int(x) for x in list('{0:07b}'.format(i))])
    labels.append(i % 2)

perceptron = Perceptron(7)
perceptron.train(training_inputs, labels)
```

```python
# 1 stands for odd and 0 stands for even

inputs = np.array([int(x) for x in list('{0:07b}'.format(5))])
print(perceptron.predict(inputs))
inputs = np.array([int(x) for x in list('{0:07b}'.format(8))])
print(perceptron.predict(inputs))
inputs = np.array([int(x) for x in list('{0:07b}'.format(1))])
print(perceptron.predict(inputs))
inputs = np.array([int(x) for x in list('{0:07b}'.format(12))])
print(perceptron.predict(inputs))
```
