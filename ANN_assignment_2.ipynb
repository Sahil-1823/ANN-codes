## AND-NOT function using Mc-Culloh Pitts model

```python
#importing neccessary libraries
import numpy as np
```

```python
def and_not(x1, x2):
    # Define the weights and threshold for the McCulloch-Pitts neuron
    weights = np.array([1, -1])
    threshold = 0.5
    
    # Calculate the weighted sum of inputs
    weighted_sum = np.dot(weights, [x1,x2])
    
    # Apply the step function to get the output
    if weighted_sum > threshold:
        return 1
    else:
        return 0
```

```python
# Test the AND NOT function
print(f"Activation : {and_not(0, 0)}") # 0 AND NOT 0 -> 0
print(f"Activation : {and_not(0, 1)}") # 0 AND NOT 1 -> 0
print(f"Activation : {and_not(1, 0)}") # 1 AND NOT 0 -> 1
print(f"Activation : {and_not(1, 1)}") # 1 AND NOT 1 -> 0
```
