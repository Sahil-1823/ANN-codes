```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

1) The linear activation function, also known as the identity function, is one of 
the simplest activation functions used in neural networks. It is defined as 
f(x) = x, where the output value of the neuron is equal to its input value. 
In other words, it performs a linear mapping of the input to the output without 
any non-linearity.

When a linear activation function is used, the output of a neuron is directly
proportional to the weighted sum of its inputs, without introducing any 
additional complexity or non-linear transformations. As a result, the neural 
network using linear activation functions is equivalent to a linear model, 
regardless of the number of layers.


```python
def linear(x):
    return x

x = np.linspace(-7, 1)  
plt.plot(x, linear(x))
plt.title("linear activation function")
plt.show
```

2) The **hyperbolic tangent** activation function, commonly known as tanh*, is a non-linear activation function that maps the input to a range between -1 and 1. It is an extension of the sigmoid function and exhibits similar characteristics but with an output range that is symmetric around zero


---

Mathematically, the tanh function is defined as follows:

f(x) = (e^x - e^(-x)) / (e^x + e^(-x))

where e is the base of the natural logarithm (approximately 2.718).

---

The tanh function has the following properties:

***Non-linearity***: Like the sigmoid function, the tanh function is non-linear, allowing neural networks to capture and learn complex relationships between inputs and outputs.

***Range***: The output of the tanh function is always between -1 and 1. As the input approaches positive or negative infinity, the output asymptotically approaches 1 and -1, respectively.

***Symmetry***: The tanh function is symmetric around the origin (x = 0). This property can be beneficial in certain scenarios where positive and negative inputs should have similar effects.

*Zero-centered*: Unlike the sigmoid function, which is centered around 0.5, the tanh function is centered around zero. This zero-centered property can make learning easier in some cases.


```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.title("tanh activation function")
plt.show
```

3) The sigmoid activation function, also known as the logistic function, is a popular non-linear activation function used in neural networks. It maps the input to a range between 0 and 1, allowing it to produce a probability-like output.

---

Mathematically, the sigmoid function is defined as:

f(x) = 1 / (1 + e^(-x))

where e is the base of the natural logarithm (approximately 2.718).

---


The sigmoid activation function possesses the following properties:

***Non-linearity***: The sigmoid function introduces non-linearity to the network, enabling it to model and learn complex relationships between inputs and outputs. This non-linearity is crucial for neural networks to handle tasks such as classification and pattern recognition.

***Range***: The output of the sigmoid function is always between 0 and 1. As the input approaches positive or negative infinity, the output asymptotically approaches 1 and 0, respectively. This property allows the sigmoid function to produce a probability-like interpretation, where values closer to 1 represent a higher probability and values closer to 0 represent a lower probability.

*Smoothness*: The sigmoid function is a smooth and continuous function, which ensures that small changes in the input result in smooth and gradual changes in the output. This property is desirable for training neural networks using gradient-based optimization algorithms.


```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.title("sigmoid activation function")
plt.show
```

4) The Rectified Linear Unit (ReLU) activation function is a popular non-linear activation function used in neural networks. It is known for its simplicity and effectiveness, and it has become one of the most widely used activation functions in deep learning.

---

The ReLU function is defined as follows:

f(x) = max(0, x)

In other words, if the input value is positive, the ReLU function returns the input value itself, and if the input value is negative, it returns 0. This means that ReLU effectively "turns on" the neuron for positive inputs, allowing information to pass through, while "turning off" the neuron for negative inputs, effectively blocking any information flow.

---

The ReLU activation function possesses several important properties:

***Non-linearity***: Like other activation functions, ReLU introduces non-linearity to the neural network, enabling it to learn complex relationships and make the network more expressive. Non-linearity is crucial for capturing and modeling non-linear patterns in the data.

***Sparsity***: ReLU can produce sparse activation, meaning that only a subset of neurons are activated while others are deactivated. This sparsity can improve computational efficiency by reducing the overall number of active neurons and parameter updates during training.

***Avoiding the vanishing gradient problem***: ReLU mitigates the vanishing gradient problem that occurs with activation functions like sigmoid or tanh. Since ReLU only "turns on" for positive inputs, the gradient for positive inputs remains constant and does not vanish as the input increases, enabling more efficient gradient propagation and learning in deep networks.

***Efficiency***: ReLU is computationally efficient to compute compared to other activation functions like sigmoid or tanh, as it involves simple thresholding operations.

```python
def Relu(x):
    return np.maximum(0.0, x)

x = np.linspace(-10, 10)
plt.plot(x, Relu(x))
plt.title("Relu activation function")
plt.show
```
